```python
import json
import os

# Storage and configuration
contacts = {}
filename = "my_contacts.json"

def validate_phone(p):
    # Basic check: only numbers and a reasonable length
    return p.isdigit() and len(p) >= 7

def validate_email(e):
    # Checks for the basics: an @ and a dot
    return "@" in e and "." in e

def add_contact():
    print("\n--- New Contact Entry ---")
    name = input("Name: ").strip()

    if name in contacts:
        print("Wait, that name already exists!")
        return

    phone = input("Phone: ")
    if not validate_phone(phone):
        print("Error: That doesn't look like a valid phone number.")
        return

    email = input("Email: ")
    if not validate_email(email):
        print("Error: Invalid email format.")
        return

    category = input("Category (Work/Home/etc): ")

    # Store data in a nested dictionary structure
    contacts[name] = {
        "phone": phone,
        "email": email,
        "group": category
    }
    print(f"Done! {name} has been added.")

def search_contact():
    print("\n--- Search ---")
    query = input("Enter name to search: ").lower()
    found = False

    for name, info in contacts.items():
        if query in name.lower(): # Allows for partial name matches
            print(f"Found: {name} | Phone: {info['phone']} | Email: {info['email']} | Group: {info['group']}")
            found = True

    if not found:
        print("No results found for that search.")

def update_contact():
    name = input("\nWhich contact do you want to change? ")
    if name not in contacts:
        print("Can't find that person.")
        return

    print("Enter new details (leave blank to keep current value)")
    new_phone = input(f"Current Phone ({contacts[name]['phone']}): ")
    new_email = input(f"Current Email ({contacts[name]['email']}): ")

    # Only update if the user actually typed something
    if new_phone.strip():
        contacts[name]['phone'] = new_phone
    if new_email.strip():
        contacts[name]['email'] = new_email

    print("Updated successfully.")

def delete_contact():
    name = input("\nEnter name to delete: ")
    if name in contacts:
        sure = input(f"Are you 100% sure you want to delete {name}? (y/n): ")
        if sure.lower() == 'y':
            contacts.pop(name)
            print("Deleted.")
    else:
        print("Contact not found.")

def show_all():
    if not contacts:
        print("\nYour contact list is empty right now.")
        return

    # Formatting the output into a readable table
    print(f"\n{'NAME':<15} | {'PHONE':<12} | {'EMAIL':<20} | {'GROUP'}")
    print("-" * 60)
    for name in sorted(contacts.keys()):
        c = contacts[name]
        print(f"{name:<15} | {c['phone']:<12} | {c['email']:<20} | {c['group']}")

def save_data():
    # Save main JSON data
    with open(filename, 'w') as f:
        json.dump(contacts, f)
    # Simple text-based backup
    with open("backup_contacts.txt", 'w') as f:
        f.write(str(contacts))
    print("Everything saved to file.")

def get_stats():
    total = len(contacts)
    print(f"\n--- Quick Stats ---")
    print(f"You have {total} contacts in your list.")

def main():
    global contacts
    # Check if file exists before trying to load it
    if os.path.exists(filename):
        try:
            with open(filename, 'r') as f:
                contacts = json.load(f)
            print("Successfully loaded saved contacts.")
        except:
            print("Could not load file. Starting with a fresh list.")

    while True:
        print("\n--- CONTACTS MANAGER MENU ---")
        print("1. View All")
        print("2. Add Contact")
        print("3. Search")
        print("4. Update")
        print("5. Delete")
        print("6. Show Stats")
        print("7. Exit & Save")

        move = input("What would you like to do? ")

        if move == '1': show_all()
        elif move == '2': add_contact()
        elif move == '3': search_contact()
        elif move == '4': update_contact()
        elif move == '5': delete_contact()
        elif move == '6': get_stats()
        elif move == '7':
            save_data()
            print("Bye!")
            break
        else:
            print("Oops, that's not a valid option.")

if __name__ == "__main__":
    main()
```
